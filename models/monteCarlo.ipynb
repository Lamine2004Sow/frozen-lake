## Formule de mise à jour Monte Carlo

$$Q(s_t, a_t) \leftarrow Q(s_t, a_t) + \alpha \Big( G_t - Q(s_t, a_t) \Big)$$

où le gain est calculé à rebours :

$$G_t = \sum_{k=t+1}^{T} \gamma^{k-t-1} r_k = r_{t+1} + \gamma r_{t+2} + \gamma^2 r_{t+3} + \ldots$$

**Définition des éléments :**

| Symbole | Nom | Description |
|---|---|---|
| $s$ | État | Position de l'agent dans l'environnement |
| $a$ | Action | Mouvement choisi par l'agent |
| $s_t$ | État au temps $t$ | Position de l'agent à l'instant $t$ |
| $a_t$ | Action au temps $t$ | Mouvement choisi à l'instant $t$ |
| $Q(s_t, a_t)$ | Q-valeur courante | Estimation du gain depuis $s_t$ en prenant $a_t$ |
| $\alpha$ | Learning rate | Contrôle la vitesse d'apprentissage $\alpha \in [0, 1]$ |
| $G_t$ | Gain réel | Somme des rewards actualisés de $t$ à $T$ |
| $G_t - Q(s_t, a_t)$ | Erreur TD | Différence entre gain réel et gain estimé |
| $r_{t+1}$ | Reward immédiat | Récompense reçue après $a_t$ depuis $s_t$ |
| $\gamma$ | Discount factor | Pondération du futur $\gamma \in [0, 1]$ |
| $T$ | Fin d'épisode | Dernier step de la trajectoire |

```python
def calcul_gain_t(t, T, gamma, rewards):
    return sum((gamma ** (k-t-1)) * rewards[k] for k in range(t+1, T+1))
   
```

```python
# test gain a l'instant t
rewards = [0, 0, 0, 1]
gamma = 0.9
t = 2
T = 3
gain = calcul_gain_t(t, T, gamma, rewards)
assert gain != 0.9, f"Expected gain to be different from 0.9, but got {gain}"
```
